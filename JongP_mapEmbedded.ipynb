## Embeddedness calculation of the Map Equation outcome

```python
import collections
import networkx as nx
nx.__version__
```

```python
fh = open("USAirport500Opsahl.txt", 'rb')
G_nx = nx.read_weighted_edgelist("USAirport500Opsahl.txt", create_using=nx.DiGraph(), nodetype=int)
print(nx.info(G_nx))
```

```python
import numpy as np
import pandas as pd
#Inputs are cummunities & G_nx (NetworkX)

# Embeddedness score algorithm for directed networks
df = pd.DataFrame(columns=['embeddedness'])
distribution = np.array([])
for cluster in range(len(communities)):
    for node in communities[cluster]:
        if G_nx.has_node(node):
            kInt = 0
            for neighbor in G_nx.successors(node):
                if float(neighbor) in communities[cluster]:
                    kInt += 1
            outdeg = G_nx.out_degree(node)
            if outdeg != 0:
                embeddedness = float("{0:.1f}".format(float(kInt)/outdeg)) # 2f
                distribution = np.append(distribution, embeddedness)
                df = df.append({'embeddedness' : embeddedness} , ignore_index=True)
        else:
            print "EXCEPT: node", node, "not found."
```
